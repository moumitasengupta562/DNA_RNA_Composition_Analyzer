```python
# DNA/RNA Composition Analyzer in Jupyter Notebook
import pandas as pd
import matplotlib.pyplot as plt

# Example DNA/RNA sequence
sequence = "ATGCGTACGTAGCTAGCTAGCTAGCGTACGATCG"

# Convert sequence to uppercase and remove newlines/spaces
sequence = sequence.replace("\n", "").replace(" ", "").upper()

# Function to calculate nucleotide counts and percentages
def nucleotide_composition(seq):
    counts = {
        "A": seq.count("A"),
        "T": seq.count("T") if "T" in seq else 0,
        "U": seq.count("U") if "U" in seq else 0,
        "G": seq.count("G"),
        "C": seq.count("C")
    }
    total = sum(counts.values())
    percentages = {k: round(v/total*100, 2) for k, v in counts.items() if v > 0}
    return counts, percentages

# Get counts and percentages
counts, percentages = nucleotide_composition(sequence)

# Print results
print("Nucleotide Counts:", counts)
print("Nucleotide Percentages:", percentages)

# Plotting the composition
df = pd.DataFrame(list(percentages.items()), columns=["Nucleotide", "Percentage"])
plt.figure(figsize=(6,4))
plt.bar(df["Nucleotide"], df["Percentage"], color=["#FF9999","#66B2FF","#99FF99","#FFCC99","#C2C2F0"])
plt.title("Nucleotide Composition of DNA/RNA Sequence")
plt.ylabel("Percentage (%)")
plt.show()
```
